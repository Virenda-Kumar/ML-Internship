# Student Management System

```python
std = []  #List of Objects of Students 
class Student:
    name = ""
    roll_no = 0
    age = 0
    grade = ''
    batch = ''
    
    def __init__(this, name, age, roll_no, grade, batch):
        this.name =  name
        this.age = age
        this.roll_no = roll_no
        this.grade = grade
        this.batch = batch
        
        
    def display_details(this):
        print("Name: ", this.name)
        print("Age: ", this.age)
        print("Roll_No.: ", this.roll_no)
        print("Grade: ", this.grade)
        print("Batch: ", this.batch, "\n")
        
        
    def search_student(this, roll):
        for i in std:
            if len(std)==0:
                print("Student Details are EMPTY")
                break
            if(i.roll_no==roll):
                print("Details of roll number ", roll, " are: ")
                i.display_details()
                return i
        print("Roll no. ", roll," doesn't exist")
        return 0
    
    
    def update_details(this, roll):
        x = std[0].search_student(roll)
        if x!=0:
            print("Choose the details to be Updated: ")
            print("1.Name \n2.Batch \n3.Grade \n4.Age \n5.Roll no.")
            ch = int(input())
            match ch:
                case 1:
                    x.name = input("Enter new Name: ")
                case 2:
                    x.batch = input("Enter new Batch: ")
                case 3:
                    x.grade = input("Enter new Grade: ")
                case 4:
                    x.age = input("Enter new Age: ")
                case 5:
                    x.roll_no = input("Enter new Roll no.: ")
    
         
    def delete_student(this, roll):
        idx = 0
        for i in std:
            if i.roll_no==roll:
                del std[idx]
                print("Roll Number ", roll, " DELETED")
                break
            idx=idx+1
        
        
```

```python
#Switch Statement:
while 1:
    print("\n.............................................................................................")
    print("Enter  your choice from the following: ")
    print("1. Add Student \n2. Display Students \n3. Search Student \n4. Delete Student \n5. Update Details\n6. Halt Program")
    choice = int(input("===========>> "))
    print(".............................................................................................")
    

    match choice:
        case 1:
            print("\nFor a Student Enter: ")
            name = input("Name: ")
            age = int(input("Age: "))
            roll_no = int(input("Roll_No.: "))
            grade = input("Grade: ")
            batch = input("Batch: ")
            std.append(Student(name, age, roll_no, grade, batch))
            
        case 2:
            if len(std)==0:
                print("\nStudent Details are EMPTY\n")
                
            else: 
                for i in range(0,len(std)):
                    print("\nDetails of Student ", i+1, " are:")
                    std[i].display_details()
            
        case 3:
            if len(std)==0:
                print("\nStudent Details are EMPTY\n")
            else: 
                roll = int(input("\nEnter Roll number: "))
                std[0].search_student(roll)
            
        case 4:
            if len(std)==0:
                print("\nStudent Details are EMPTY\n")
                
            else: 
                roll=int(input("\nEnter Roll Number of Student to be Deleted: "))
                std[0].delete_student(roll)
        
        case 5:
            roll=int(input("\nEnter Roll Number of Student to be Updated: "))
            std[0].update_details(roll)
            
        case 6:
            break
        
            
        case default:
            print("\nEnter Valid Choice!")
        
```
